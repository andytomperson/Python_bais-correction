# 把辅助数据转置，并把0的行去掉

```python
import os
import pandas as pd

# 设置文件夹路径
folder_path = 'E:/2.全球数据分区/2024010500.CA-1站点的辅助变量'

# 遍历文件夹中的所有 CSV 文件
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        # 构建完整文件路径
        file_path = os.path.join(folder_path, filename)

        # 读取 CSV 文件
        df = pd.read_csv(file_path)

        # 转置 DataFrame
        df_transposed = df.transpose()

        # 设置列名为'date'和'value'
        df_transposed.columns = [ 'value']

        # 删去倒数4行和value列为0的行
        df_filtered = df_transposed.iloc[:-4].loc[df_transposed['value'] != 0].dropna(subset=['value'])

        # 将结果保存为新的 CSV 文件，文件名加上'_processed'后缀
        output_filename = filename.replace('.csv', '_processed.csv')
        output_file_path = os.path.join(folder_path, output_filename)
        df_filtered.to_csv(output_file_path, index=False)

```

```python
import ee

# 初始化 Earth Engine
ee.Initialize()

# 定义重采样的空间分辨率
resampling_resolution = 500

# 定义重采样函数
def resample_image(image):
    # 获取原始图像的投影信息
    projection_info = image.projection()

    # 进行重采样
    resampled_image = image.reproject(crs=projection_info.crs(),
                                      scale=resampling_resolution)
    
    return resampled_image

# 加载图像集
dataset_aqy = ee.ImageCollection('NASA_USDA/HSL/SMAP10KM_soil_moisture') \
                   .select('ssm') \
                   .filter(ee.Filter.date('1999-01-01', '2022-12-31'))

# 对图像集中的每个图像应用重采样函数
resampled_images = dataset_aqy.map(resample_image)

# 定义输出CSV文件路径
out_csv = os.path.join(work_dir, 'world_SMAP_soil_moisture_2019-2020.csv')

# 将重采样后的图像集输出为CSV文件
geemap.extract_values_to_points(in_fc, resampled_images, out_csv)
```

```python
resampling_resolution = 500

# 定义重采样函数
def resample_image(image):
    # 获取原始图像的投影信息
    projection_info = image.projection()

    # 进行重采样
    resampled_image = image.reproject(crs=projection_info.crs(),
                                      scale=resampling_resolution)
    
    return resampled_image

# 加载图像集
dataset_aqy = ee.ImageCollection('NASA_USDA/HSL/SMAP10KM_soil_moisture') \
                   .select('ssm') \
                   .filter(ee.Filter.date('2015-01-01', '2015-12-31'))

# 对图像集中的每个图像应用重采样函数
resampled_images = dataset_aqy.map(resample_image) \
                              .toBands()

# # 定义输出CSV文件路径
out_csv = os.path.join(work_dir, 'CA-1_SMAP_soil_moisture_2015-2015.csv')

# # 将重采样后的图像集输出为CSV文件
geemap.extract_values_to_points(in_fc, resampled_images, out_csv)
```
